# read file and make it in one data fram

```python
import numpy as np
import pandas as pd 
import os
```

```python
files = [file for file in os.listdir('./Sales_Data')]
data = pd.DataFrame()
for file in files:
    df =pd.read_csv("./Sales_Data/"+file)
    data = pd.concat([data, df])
    data.to_csv(index=False)
```

```python
data
```

# Delete Rows don't have a value

```python
data[data['Price Each'].isna()]
```

```python
data = data.dropna()
```

```python
data.loc[:, 'Price Each'] = pd.to_numeric(data["Price Each"], errors='coerce')
```

```python
data[data['Price Each'].isna()]
```

```python
data=data.dropna()
```

# convert order Data to datetime type

```python
data["Order Date"].dtype
```

```python
data.loc[:, "Order Date"] = pd.to_datetime(data.loc[:, "Order Date"])
```

```python
data["Order Date"]
```

```python
data
```

# Add col month

```python
data.loc[:, 'month'] = data.loc[:, "Order Date"].dt.month
```

```python
data
```

```python
data = pd.merge(data, data.groupby('month')['Price Each'].sum(), on='month', suffixes=('', '_MonthlySum'))
```

```python
data
```

```python
data['month'] = data['month'].astype(int)
```

```python
data['month'].dtype
```

```python
data['Quantity Ordered']=data['Quantity Ordered'].astype(int)
```

```python
data.groupby('month')[['Quantity Ordered', 'Price Each_MonthlySum']].sum()
```

# Add col city

```python
def get_city(address):
    return address.split(",")[1].strip(" ")

def get_state(address):
    return address.split(",")[2].split(" ")[1]
data['city'] = data['Purchase Address'].apply(lambda x: f"{get_city(x)}  ({get_state(x)})")
```

```python
data
```

# find city with the highest sales

```python
data.groupby('city')[['Quantity Ordered', 'Price Each', 'Price Each_MonthlySum']].sum()
```
